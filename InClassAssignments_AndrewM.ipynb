A company wants to calculate its Gross Profit as: Sales - Costs of Goods Sold


*   assume that cost of goods sold is 40%

The company wants to calclate taxes and net income:


*   prompt for a tax rate in %
*   To calculate taxes, multiply gross profit by tax rate, and to get net income you subtract taxes from gross profit





```python
Sales = input(("Enter Sales ($): "))
Sales = float(Sales)

cogs_pct = input("Enter Cost of Goods (%): ")
cogs_pct = float(cogs_pct)

cogs = Sales * (cogs_pct/100)

Gross_Profit = Sales - cogs

tax_rate = input("Enter Tax Rate (%): ")

tax_rate = float (tax_rate)

taxes = Gross_Profit * (tax_rate/100)

net_income = Gross_Profit - taxes
print("sales:", Sales)
print("Cost of goods:",cogs)
print("Gross Profit:", Gross_Profit)
print("Taxes:", taxes)
print("Net Income:", net_income)
```

# Inventory Management

Assume an item has a steady demand during every year. There is a cost for placing an order with the supplier and a cost for restocking this item. There is also a cost for storing items (e.g. warehouse). Given these costs and your demand forecast, calculate how much of the item to reorder to minimize annual ordering and storing costs. This amount is referred to as the Economic order Quantity (EOQ)
Given the following assumptions
D -> Annual demand (units/year)
S -> reorder cost ($/order)
H = holding costs ($/unit/year)
then EOQ = sqrt(2*D*S/H)


```python
# Inputs

demand = input("enter demand (units/year): ")
demand = int(demand)

reorder = float(input("Enter reorder cost ($/order): "))

holding = float(input("Enter holding cost ($/unit/year): "))

# Process

EOQ = (2 * demand * reorder / holding) ** (1/2)

# Output

print(f"EOQ: {EOQ} units")




```

the classic hand game. write a program that plays the game against the computer for as long as the user wants. Start by allowing the user to choose whether they want to play using the appropriate input as long as the user chooses 'yes', the game continues Add realism by allowing the computer to randomly choose its action. you can do this using the 'random' library: import random computer = random.choice(['rock','paper', 'scissors'])

```python
# AI
import random

print("Welcome to Rock, Paper, Scissors!")

# Main game loop
while True:
    play = input("Do you want to play? (yes/no): ").strip().lower()

    if play != "yes":
        print("Thanks for playing! Goodbye!")
        break

    # User choice
    user_choice = input("Choose rock, paper, or scissors: ").strip().lower()

    # Validate input
    if user_choice not in ['rock', 'paper', 'scissors']:
        print("Invalid choice! Please select rock, paper, or scissors.")
        continue

    # Computer choice
    computer_choice = random.choice(['rock', 'paper', 'scissors'])
    print(f"The computer chose: {computer_choice}")

    # Determine winner
    if user_choice == computer_choice:
        print("It's a tie!")
    elif (
        (user_choice == 'rock' and computer_choice == 'scissors') or
        (user_choice == 'paper' and computer_choice == 'rock') or
        (user_choice == 'scissors' and computer_choice == 'paper')
    ):
        print("You win!")
    else:
        print("Computer wins!")

    print("-" * 30)  # Divider for readability
```

A quality control manager would like to find the average weight of a list of weights. Prior experience indicates that the first two measures are atypical, and the practice is to exclude these from the average calculation.

Write a program that calculates the average of a list of numbers that given as inputs from the user, excluding the first two values.

For your program, you must define and use the following functions:

getValues(n): Prompt the user for n values, returns the list
calcAverage(alist): Calculates the average from the list

differenc with assignment II: use try-except to avoid invalid inputs (e.g. neg values, strings), invalid operations, etc

```python
def getValues(n: int) -> list:
  items = []

  for i in range(n):
    try:
      item = float(input('Enter weigh value: '))
    except Exception as e:
      print(e)
      item = float('nan')


    items.append(item)

  return items

def calcAverage(alist: list) -> float:

  try:
    return sum(alist[2:]) / len(alist[2:])
  except ZeroDivisionError:
    return float('nan')

def main():
  num = int(input('Enter a number of weirghts: '))
  if num <= 0:
    raise ValueError('num must be positive')

  weights = getValues(num)
  avg = calcAverage(weights)
  print(f'Average weight: {avg} lbs.')


main()
```

Greeting class

write a python class and populate it with two attributes and methods:
- Attributes: First/Last Names
- Method display(): takes a Boolean input parameter; displays the attribute values as either formally(last, first) or informally (first)
- special method __str__()


```python
class Greeting:
  def __init__(self, first, last):
    self.first = first
    self.last = last

  def __str__(self):
    return "Instance of class Greeting"

  def display(self, reverse=True):
    if reverse:
      print(f'{self.last}, {self.first}')
    else:
      print(self.first)

name = Greeting('John' , 'Doe')
print(name)
name.display()
name.display(reverse=False)
```

Create a class Point that hos:
- two coordinate attributes (x, y)
- a method distance() that calculates the distance between two instances of the class

Create a class Rectangle that has:
- six attributes: four Point instances and two attributes for the width/height. the latter two are calculated based on pairs of the four Point instances
- three methods for calculating width, height, area of the rectangle

```python
class Point:
  def __init__(self, x, y):
    self.x = x
    self.y = y


  def distance(self, other):
    x_diff_sq = (self.x - other.x)**2
    y_diff_sq = (self.y - other.y)**2
    return (x_diff_sq + y_diff_sq)**0.5

class Rectangle:
  def __init__(self):
    self.p1 = Point(0,0)
    self.p2 = Point(1,0)
    self.p3 = Point(1,1)
    self.p4 = Point(0,1)
    self.width = self.p1.distance(self.p2)
    self.height = self.p1.distance(self.p4)

  def area(self):
   return self.width * self.height


sq = Rectangle()
sq.area()



```
